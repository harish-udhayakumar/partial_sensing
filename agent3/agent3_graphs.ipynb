```python
def calculateAllValues(dim,p_low,p_high,step,num_of_runs):
    global maze_size
    probabilities = np.arange(p_low, p_high, step)
    probabilities = np.round(probabilities,3)
    trajectory_length ={}
    num_cells = {}
    avg_length = {}
    avg_length_2 = {}
    planning_time ={}
    
    print("probabilities: ",probabilities)
    for prob in probabilities:
        for i in range(num_of_runs):
            start = time.time()
            print("started for ", prob, " run: ",i)
            initialize_all_matrices(prob)
            _,_,_,path = astar((0,0,0),block_unblock_matrix)
#             print("path a*:",path)
            if len(path) == 0:
                continue
            else:
                    
                shortest_path_on_discovered_gridworld,traj_length,num_cells_processed,plan_time = agent3()
                
                if prob in trajectory_length.keys():
                    trajectory_length[prob].append(traj_length)
                else:
                    trajectory_length[prob] = list()
                    trajectory_length[prob].append(traj_length)
                
                if prob in avg_length.keys():
                    avg_length[prob].append(traj_length/len(shortest_path_on_discovered_gridworld))
                else:
                    avg_length[prob] = list()
                    avg_length[prob].append(traj_length/len(shortest_path_on_discovered_gridworld))
                    
                if prob in avg_length_2.keys():
                    avg_length_2[prob].append(len(shortest_path_on_discovered_gridworld)/len(path))
                else:
                    avg_length_2[prob] = list()
                    avg_length_2[prob].append(len(shortest_path_on_discovered_gridworld)/len(path))
                    
                if prob in num_cells.keys():
                    num_cells[prob].append(num_cells_processed)
                else:
                    num_cells[prob] = list()
                    num_cells[prob].append(num_cells_processed)
                
                if prob in planning_time.keys():
                    planning_time[prob].append(plan_time)
                else:
                    planning_time[prob] = list()
                    planning_time[prob].append(plan_time)
             
            time_taken = time.time() - start
            print("Finished for probability: ",prob," run: ",i, " time taken: ",time_taken)
        print("trajectory_length for prob: ",prob," value: ",sum(trajectory_length[prob])/len(trajectory_length[prob]))
        print("avg_length for prob: ",prob," value: ",sum(avg_length[prob])/len(avg_length[prob]))
        print("avg_length_2 for prob: ",prob," value: ",sum(avg_length_2[prob])/len(avg_length_2[prob]))
        print("num_cells for prob: ",prob," value: ",sum(num_cells[prob])/len(num_cells[prob]))
        print("planning_time for prob: ",prob," value: ",sum(planning_time[prob])/len(planning_time[prob]))
        
    #Density vs Avg Trajectory Length:
    filename = "./results/avg_traj_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in trajectory_length.keys():
        value = trajectory_length[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average (Length of Trajectory/length of shortest path on final discovered gridworld)
    filename = "./results/avg_len_1_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in avg_length.keys():
        value = avg_length[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average (Length of shortest path on final discovered gridworld / Length of shortest path in Full Gridworld)
    filename = "./results/avg_len_2_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in avg_length_2.keys():
        value = avg_length_2[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS Average Number of Cells Processed
    filename = "./results/avg_cells_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in num_cells.keys():
        value = num_cells[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
    
    #Density VS planning time
    filename = "./results/planning_time_"+str(np.random.randint(1000000))+".txt"
    f = open(filename, "w")
    f.write(str(dim)+str(",")+str(num_of_runs)+"\n")
    for key in planning_time.keys():
        value = planning_time[key]
        f.write(str(key)+ str(",")+str(sum(value)/len(value))+"\n")
    print("done. file: ",filename)
```
